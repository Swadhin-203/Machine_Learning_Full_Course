# Decission Tree Classifier Implementation With Post Prunning and Preprunning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
df=sns.load_dataset("iris")
```

```python
#Segregreating Dependent and the Indpendent Feature
X=df.iloc[:,:-1]
y=dataset.target
```

```python
X
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion="entropy")
```

```python
classifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

# Post prunning 

```python
classifier=DecisionTreeClassifier(criterion="entropy",max_depth=2)
classifier.fit(X_train,y_train)
```

```python
tree.plot_tree(classifier,filled=True)
```

```python
# Prediction
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
# Checking the accuracy of the result
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print("The accuracy score is ",score)
print(classification_report(y_pred,y_test))
```

# Decission Tree Prepruning and Hyperparameter Tuning For Huge Data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier=DecisionTreeClassifier()
clf=GridSearchCV(classifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print("The final Score is ",score)
print(classification_report(y_pred,y_test))
```
